```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
url="https://www.propertywala.com/properties/type-residential_apartment_flat/for-sale/location-hyderabad_telangana?page=10"
```

```python
Area = []
Price = []
SqFt=[]
SqFt_price=[]
Bedrooms=[]
Posted_date=[]


for i in range(1,51):
    url=f"https://www.propertywala.com/properties/type-residential_apartment_flat/for-sale/location-hyderabad_telangana?page={i}"

    page=requests.get(url)

    soup=BeautifulSoup(page.text)

    #area
    for i in soup.find_all("article",class_='Live'):
        f=i.find('h4')
        #print(f.text)
        regex=re.findall("\s\w+\s\w+,\s\w+",f.text)
        #print(regex)
        if regex:
            Area.append(regex[0])
        else:
            Area.append(np.nan)
        
            
     #price
        g = i.find('div',class_='propertyPrice')
        #print(g.text)
        if g:
            Price.append(g.text.strip())
        else:
            Price.append(np.nan)
            
            
     #sqFt
        a=i.find('span',class_='areaUnit')
        #print(a.text)
        #regex=re.findall("\d.*",a.text)
        #print(regex)
        if a:
            SqFt.append(a.text)
        else:
            SqFt.append(np.nan)
            
            
      #sqft_price
        j=i.find('li',class_='features')
        #print(j.text)
        regex=re.findall("Rs\.(\d+)\/SqFeet",j.text)
        #print(regex)
        if regex:
            SqFt_price.append(regex[0])
        else:
            SqFt_price.append(np.nan)
            
            
      #bedrooms
        m=i.find('h4')        
        #print(m.text)
        regex=re.findall("\d.Bedroom",m.text)
        #print(regex)
        if regex:
            Bedrooms.append(regex[0])
        else:
            Bedrooms.append(np.nan)
            
            
      #posted_date
        container=i.find_all("div",class_="summary")
        for j in container:
            n=i.find('li',class_="posted")
            if n:
                Posted_date.append(n.time.text)
            else:
                Posted_date.append(np.nan)
#             regex=re.findall("\d+-\d+-\d+",n.text)
#             print(regex)
#             if regex:
#                 posted_date.append(regex)
#             else:
#                 posted_date.append(np.nan)

         
        
            
               
print((Area))
print((Price))
print(SqFt)
print(SqFt_price)
print(Bedrooms)
print(Posted_date)



```

```python
print(len(Area))
print(len(Price))
print(len(SqFt))
print(len(SqFt_price))
print(len(Bedrooms))
print(len(Posted_date))
```

```python
flats={"area":Area,"price":Price,"sqFt":SqFt,"sqFt price":SqFt_price,"Bedrooms":Bedrooms,"posted date":Posted_date}
```

```python
df=pd.DataFrame(flats)
```
